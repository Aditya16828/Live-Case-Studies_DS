```python
# Wrapping all the commands in a function
def remove_na_dfc(dfc, annual_income_avg):
    dfc_copy = dfc.copy()
    left = dfc_copy[dfc_copy['annual_income'].isnull()]
    t1 = pd.merge(left, annual_income_avg, how='left', on = 'marital_status')
    t1['annual_income_x'], t1['annual_income_y'] = t1['annual_income_y'], t1['annual_income_x']
    t1 = t1.drop('annual_income_y', axis = 1)
    t1.rename(columns={'annual_income_x': 'annual_income'}, inplace=True)
    dfc_copy = pd.concat([t1, dfc_copy[~dfc_copy['annual_income'].isnull()]])
    return dfc_copy
(remove_na_dfc(dfc, annual_income_avg))
```

```python
def annual_income_avg_def1(annual_income, marital_status):
    if pd.isna(annual_income):
        return annual_income_avg[annual_income_avg['marital_status'] == marital_status]['annual_income']
    return annual_income
```

```python
dfc['annual_income'] = dfc.apply(lambda x: annual_income_avg_def1(x['annual_income'], x['marital_status']), axis = 1)
dfc
```

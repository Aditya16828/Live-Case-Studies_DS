```python
dfc['marital_status'] = dfc['marital_status'].replace('Widow', 'Widowed')
```

```python
dfc['marital_status'].unique()
```

```python
annual_income_avg = dfc.groupby(by = 'marital_status').mean()['annual_income'].reset_index()
print(annual_income_avg)
```

```python
left = dfc[dfc['annual_income'].isnull()]
t1 = pd.merge(left, annual_income_avg, how='left', on = 'marital_status')
t1
```

```python
t1['annual_income_x'], t1['annual_income_y'] = t1['annual_income_y'], t1['annual_income_x']
t1
```

```python
t1 = t1.drop('annual_income_y', axis = 1)
t1
```

```python
t1.rename(columns={'annual_income_x': 'annual_income'}, inplace=True)
type(t1)
```

```python
dfc = pd.concat([t1, dfc[~dfc['annual_income'].isnull()]])
dfc
```
